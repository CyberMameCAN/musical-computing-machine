```python
import numpy as np
import pandas as pd
import urllib.request
import matplotlib.pyplot as plt
%matplotlib inline

import sklearn #機械学習のライブラリ
from sklearn.decomposition import PCA #主成分分析

from pandas import plotting 
```

```python
url = "https://raw.githubusercontent.com/maskot1977/ipython_notebook/master/toydata/wine.txt"
urllib.request.urlretrieve(url, 'wine.txt')
```

```python
df = pd.read_csv("wine.txt", sep="\t", index_col=0)
```

```python
mean = df.mean()
sigma = df.std()
print('- 平均 ---')
print(mean)
print('- 標準偏差 ---')
print(sigma)
```

```python
org_len = len(df)
low =   mean['Magnesium'] - 3*sigma['Magnesium']
heigh = mean['Magnesium'] + 3*sigma['Magnesium']

dummy = df[(df['Magnesium']>low) & (df['Magnesium']<heigh)]
dummy_len = len(dummy)

print(f'{org_len} -> {dummy_len}')
```

```python
_df = df

cols = df.columns
for col in cols:
    low = mean[col] - 3 * sigma[col]
    high = mean[col] + 3 * sigma[col]
    
    _df = _df[(_df[col] > low) & (_df[col] < high)]
```

```python
len(_df)
```

```python
sns.boxplot(x="class", y="Magnesium", data=_df)
```

```python
# 標準化　平均値を引いたものを標準偏差で割る
dfs = df.iloc[:, 1:].apply(lambda x: (x-x.mean())/x.std(), axis=0)
dfs.head()
```

## 主成分分析

```python
pca = PCA()
pca.fit(dfs)
# データを主成分空間に写像
feature = pca.transform(dfs)
```

```python
# 主成分得点
pd.DataFrame(feature, columns=["PC{}".format(x + 1) for x in range(len(dfs.columns))]).head()
```

```python
# 第一主成分と第二主成分でプロットする
plt.figure(figsize=(6, 6))
plt.scatter(feature[:, 0], feature[:, 1], alpha=0.8, c=list(df.iloc[:, 0]))
plt.grid()
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.show()
```

第一主成分と第二主成分だけで、良い感じで分けられていそう

```python
plotting.scatter_matrix(pd.DataFrame(feature, 
                        columns=["PC{}".format(x + 1) for x in range(len(dfs.columns))]), 
                        figsize=(12, 12),
                        c=list(df.iloc[:, 0]),
                        alpha=0.5) 
plt.show()
```

## 寄与率と累積寄与率

```python
# 寄与率
pd.DataFrame(pca.explained_variance_ratio_, index=["PC{}".format(x + 1) for x in range(len(dfs.columns))])
```

```python
# 累積寄与率を図示する
import matplotlib.ticker as ticker
plt.gca().get_xaxis().set_major_locator(ticker.MaxNLocator(integer=True))
plt.plot([0] + list( np.cumsum(pca.explained_variance_ratio_)), "-o")
plt.xlabel("Number of principal components")
plt.ylabel("Cumulative contribution rate")
plt.grid()
plt.show()
```

## 固有値と固有ベクトル

```python
# PCA の固有値
pd.DataFrame(pca.explained_variance_, index=["PC{}".format(x + 1) for x in range(len(dfs.columns))])
```

```python
# PCA の固有ベクトル
pd.DataFrame(pca.components_, columns=df.columns[1:], index=["PC{}".format(x + 1) for x in range(len(dfs.columns))])
```

```python
# 第一主成分と第二主成分における観測変数の寄与度をプロットする
plt.figure(figsize=(8, 8))
for x, y, name in zip(pca.components_[0], pca.components_[1], df.columns[1:]):
    plt.text(x, y, name)
plt.scatter(pca.components_[0], pca.components_[1], alpha=0.8)
plt.grid()
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.show()
```
