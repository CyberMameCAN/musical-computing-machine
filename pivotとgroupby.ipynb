```python
import sys
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
 
#from IPython.display import set_matplotlib_formats
from matplotlib_inline.backend_inline import set_matplotlib_formats # バージョンによってはこちらを有効に
set_matplotlib_formats('retina')
 
# CSV ファイルを読み込む
url = "https://github.com/rinsaka/sample-data-sets/blob/master/machine-sales.csv?raw=true"
msales = pd.read_csv(url, parse_dates=["date"])
msales.info()
```

```python
msales
```

```python
msales['year']  = msales['date'].dt.strftime('%Y')
msales['month'] = msales['date'].dt.strftime('%m')
msales['day']   = msales['date'].dt.strftime('%d')
```

```python
msales
```

## pivot_table

```python
staff_summary = pd.pivot_table(msales, index='staff', columns='year', values='amount', aggfunc='sum').reset_index()
staff_summary
```

```python
pd.pivot_table(msales, index='staff', columns='year', values='amount', aggfunc=['sum', 'mean']).reset_index()
```

```python
pd.pivot_table(msales, index=['year', 'month'], columns='staff', values='amount', aggfunc=['sum', 'mean'])
```
